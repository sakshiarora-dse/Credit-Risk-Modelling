## Data preparation

### Dependent Variable. Good/ Bad (Default) Definition. Default and Non-default Accounts.

- Loans that have defaulted will be called bad, while those which performed well will be called good, therefore, need to distinguish between good and bad loans
- Typically, default definitions are based on the delinquency of the borrower measured in days past the payment due date.
- A common definition is that the borrower has defaulted if they are more than 90 days past due on a loan.
- Usually we create a new variable in the data set of a boolean or binary type i.e. zero will stand for a default or bad loan, while one for a good loan.

### Logistic Regression

- The established statistical methodology to model probability of default is a logistic regression where the dependent variable is precisely whether a customer defaulted or not.
- Logistic regression estimates the relationship between two things, the logarithm of odds of an outcome of interest or dependent variable and a linear combination of predictors or independent variables.
- In our case, the outcome of interest is the non default or default event. So on the left hand side, we have the logarithm of the ratio of the non defaults to defaults or the odds ratio.
- The established practice is for all independent variables in the probability of default models to be dummy variables. That is binary categorical variables or indicator variables. In other words, we'll have to include dummy variables as independent variables in our model.

### Preprocessing Discrete Variables: Automating Calculaions

```python
# WoE function for discrete unordered variables
def woe_discrete(df, discrete_variabe_name, good_bad_variable_df):
    df = pd.concat([df[discrete_variabe_name], good_bad_variable_df], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                    df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[:, [0, 1, 3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['n_bad'] = (1 - df['prop_good']) * df['n_obs']
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum()
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad'])
    df = df.sort_values(['WoE'])
    df = df.reset_index(drop = True)
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    return df
# Here we combine all of the operations above in a function.
# The function takes 3 arguments: a dataframe, a string, and a dataframe. The function returns a dataframe as a result.
```
